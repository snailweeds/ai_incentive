```python
f = open("seoul.csv", "r", encoding="cp949")
data = csv.reader(f)
header = next(data)
max_temp = -999
max_data = ""

for row in data:
    if row[-1] == "":
        row[-1] = -999
    row[-1] = float(row[-1]) # 문자로 인식된 숫자를 숫자로 변경
    if max_temp < row[-1]:
        max_data = row[0]
        max_temp = row[-1]
f.close()
print("기상 관측 이래 서울 최고 기온이 가장 높았던 날은", max_data+"로, ", max_temp,"도 였습니다.")
```

```python
# 생일날의 날씨
f = open("seoul.csv")
data = csv.reader(f)
next(data)
high = []
low = []
for row in data:
    if row[-1] != "" and row[-2] != "":
        if 1983 <= int(row[0].split("-")[0]):
            if row[0].split("-")[1] == "02" and row[0].split("-")[2] == "14":
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.plot(high, "pink")
plt.plot(low, "skyblue")
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('내 생일의 기온 변화 그래프')
plt.plot(high, 'hotpink', label = 'high')
plt.legend()
plt.show()
```

```python
# 생일날의 날씨
f = open("seoul.csv")
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split("-")[1]
    if row[-1] != "":
        if month == "08":
            aug.append(float(row[-1]))
        if month == "01":
            jan.append(float(row[-1]))
plt.hist(aug, bins = 100, color = "r", label = "Aug")
plt.hist(jan, bins = 100, color = "b", label = "Jan")
plt.legend()
plt.show()
```
